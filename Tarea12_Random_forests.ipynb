```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

#### División de los datos

```python
#Cargar los datos desde repositorio online:
da = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data',delimiter=',',header=None)
da.columns = ["Sex","Length","Diameter","Height","Whole_weight",
                  "Shucked_weight","Viscera_weight","Shell_weight","Rings"]

da.Sex = da.Sex.astype('category').cat.codes
da.Length = da.Length.astype('category').cat.codes
da.Diameter = da.Diameter.astype('category').cat.codes
da.Height = da.Height.astype('category').cat.codes
da.Whole_weight = da.Whole_weight.astype('category').cat.codes
da.Shucked_weight = da.Shucked_weight.astype('category').cat.codes
da.Viscera_weight = da.Viscera_weight.astype('category').cat.codes
da.Shell_weight = da.Shell_weight.astype('category').cat.codes
da.Rings = da.Rings.astype('category').cat.codes

Y = da["Rings"]
X = da.drop('Rings', 1)
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size=0.20, random_state=136894)

n = 100
```

#### Decision tree:

```python
dts = np.zeros(n)
for i in range(0,n):
    clf = DecisionTreeClassifier()
    clf = clf.fit(xtrain, ytrain)
    y_pred = clf.predict(xtest)
    dts[i] = accuracy_score(ytest, y_pred)
```

#### Random Forest:

```python
rfs = np.zeros(n)
for i in range(0,n):
    clfR = RandomForestClassifier()
    clfR = clfR.fit(xtrain, ytrain)
    y_predR = clfR.predict(xtest)
    rfs[i] = accuracy_score(ytest, y_predR)
```

### Comparativo de ambos algoritmos

```python
plt.plot(range(0,n),rfs)
plt.plot(range(0,n),dts)
plt.legend(["Random forest", "Decision tree"])
plt.show()
```

### Conclusiones:
Es evidente que el algoritmo de Random Forest presenta un mejor desempeño casi todas las veces.
